# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the test set results
No need to change the plt.plot variables from X_train -> X_test as linear model (same line will be output in both cases)

```python
plt.scatter(X_test, y_test, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Making single prediction (for example salary of an employee with 12 years of experience) 

```python
print(regressor.predict([[12]]))
```

Our model predicts that the salary of an employee with 12 years of experience is $138,531.00

**Note**: Notice that the value of the feature (12 years) was input in a double pair of square brackets, that's because the predict model always expects a 2D array as the format of the inputs and putting 12 in the double pair of square brackets makes the input exactly a 2D array. Simply put

12 --> scalar
</br>
[12] --> 1D array
</br>
[[12]] --> 2D array

Getting the final linear regression equation with the values of the coefficients

```python
print(regressor.coef_)
print(regressor.intercept_)
```

##### Therefore the equation of our simple linear regression model is :
</br>
