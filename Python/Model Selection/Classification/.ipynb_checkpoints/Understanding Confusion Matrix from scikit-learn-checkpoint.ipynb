The confusion matrix is a summary table used to evaluate the performance of a classification model. It shows the counts of predicted and actual classes, providing insights into how well your model is performing. It's especially useful for binary and multiclass classification problems.

In scikit-learn, the confusion matrix can be generated using the confusion_matrix function:

```python
from sklearn.metrics import confusion_matrix

# Example
y_true = [0, 1, 1, 0, 1, 0, 1, 1, 0]  # Ground truth labels
y_pred = [0, 1, 0, 0, 1, 0, 1, 1, 1]  # Predicted labels

cm = confusion_matrix(y_true, y_pred)
print(cm)
```

### Structure of the confusion matrix

For a binary classification problem.

Where:
- **True Negative (TN):** Model predicted 0 (negative) when the actual class is 0.
- **False Positive (FP):** Model predicted 1 (positive) when the actual class is 0 (Type I error).
- **False Negative (FN):** Model predicted 0 (negative) when the actual class is 1 (Type II error).
- **True Positive (TP):** Model predicted 1 (positive) when the actual class is 1.

### Key metrics from confusion matrix

From the confusion matrix, you can compute the following metrics to evaluate the model:

1. **Accuracy:** Overall correctness of the model.<br><br>
                    Accurracy = (TP + TN) / (Total Samples)<br>
   - **Use:** 
       1. Gives a quick understanding of how well the model performs overall.
       1. Works well when the dataset is balanced (i.e., equal distribution of classes).<br><br>
       
   - **Limitations:** Misleading in cases of class imbalance. For example, if 95% of the data belongs to one class, a model predicting only that class will have 95% accuracy but won't perform well.<br><br><br>
 
1. **Precision:** Out of the predicted positive samples, how many are actually positive.<br><br>
            Precision = TP / (TP + FP)<br>
   - **Use:**
       1. Useful in scenarios where **false positives** are costly.
       1. Example: In spam detection, falsely marking important emails as spam (false positives) is undesirable.<br><br>
       
   - **Focus:** Ensures that when the model predicts a positive, itâ€™s highly likely to be correct.<br><br><br>
            
1. **Recall (Sensitivity/True Positive Rate):** Out of the actual positive samples, how many did the model correctly predict as positive.<br><br>
            Recall = TP / (TP + FN)<br>
   - **Use:**
       1. Useful in scenarios where **false negatives** are costly.
       1. Example: In medical diagnoses, missing a positive case (false negative) can lead to severe consequences.<br><br>
       
   - **Focus:** Maximizes detection of true positives, even at the cost of more false positives.<br><br><br>
            
1. **F1 Score:** Harmonic mean of precision and recall.<br><br>
            F1 Score = 2 * ((Precision * Recall) / (Precision + Recall))<br>
   - **Use:**
       1. Best for **imbalanced datasets**, where accuracy can be misleading.
       1. Useful when precision and recall are equally important.<br><br>
       
   - **Focus:** A single score that reflects the trade-off between precision and recall.<br><br><br>
            
1. **Specificity (True Negative Rate):** Out of the actual negative samples, how many did the model correctly predict as negative.<br><br>
            Specificity = TN / (TN + FP)<br>
   - **Use:**
       1. Useful when detecting negatives accurately is important.
       1. Example: In fraud detection, ensuring legitimate transactions are not flagged as fraud (false positives) is critical.<br><br>
       
   - **Focus:** Complements recall by focusing on the correct identification of negatives.<br><br><br>

### Example output

For the given y_true and y_pred in the example:

Confusion Matrix:<br>
[[3 1]<br>
&nbsp;[1 4]]

- **TN** = 3
- **FP** = 1
- **FN** = 1
- **TP** = 4

From this:
- Accurracy = (3 + 4) / 9 = 0.78 <br>
- Precision = 4 / (4 + 1) = 0.8 <br>
- Recall = 4 / (4 + 1) = 0.8 <br>
- F1 Score = 2 * ((0.8 * 0.8) / (0.8 = 0.8)) = 0.8 <br>
- Specificity = 3 / (3 + 1) = 0.75 <br>

### Visualizing the Confusion Matrix

Scikit-learn provides ConfusionMatrixDisplay for visualizing the confusion matrix:

```python
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot()
```

### Why Use Different Metrics?

- **Task-Specific Needs:** Different applications prioritize different errors:
    - **Medical Testing:** High recall (sensitivity) is critical to catch all positive cases.
    - **Spam Detection:"" High precision is important to avoid labeling legitimate emails as spam.
- **Imbalanced Datasets:** Metrics like precision, recall, and F1 score give a clearer picture when one class dominates.
- **Decision Making:** Metrics guide hyperparameter tuning, model selection, and thresholds for classification.
